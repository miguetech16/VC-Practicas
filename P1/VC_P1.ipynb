```python
vid = cv2.VideoCapture(0)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Separamos canales
        b = frame[:,:,0]
        g = frame[:,:,1]
        r = frame[:,:,2]
        #Dimensiones imagen de entrada
        h, w, c = frame.shape

        # MODIFICACIONES EN LOS DIFERENTES PLANOS 

        # Aumentar el contraste del plano rojo
        r = cv2.convertScaleAbs(r, alpha=2.0, beta=0)

        # Aplicar detección de bordes (Canny) al plano verde
        g = cv2.Canny(g, 100, 200)

        # Aplicar un filtro de umbralización al plano azul
        _, b = cv2.threshold(b, 127, 255, cv2.THRESH_BINARY)

        #Concateamos en horizontal los tres planos del fotograma
        collage = np.hstack((r, g, b))
    
        # Muestra fotograma redimensionando a la mitad para que quepa en pantalla
        cv2.imshow('RGB', cv2.resize(collage, (int(w*1.5),int(h/2)),cv2.INTER_NEAREST))
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Evento de ratón para mostrar el valor RGB del píxel

```python
# Función que trata eventos de ratón 
def mouse_events(event, x, y, flags, params):
    global px, py, R, G, B
    px = -1
    # Movimiento del puntero     
    if event==cv2.EVENT_MOUSEMOVE:
        #Valores del píxel, desde la webcam no es BGR
        B = frame[y, x, 0]
        G = frame[y, x, 1]
        R = frame[y, x, 2]  
        # Almacena valores del evento
        px = x
        py = y
```

```python
vid = cv2.VideoCapture(0)

# Tipografía para mostrar texto
font = cv2.FONT_HERSHEY_SIMPLEX
  
lanzado = 0 #bandera para primer fotograma
px = -1
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret: 
        # Activa el manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)   
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1
            
        # Muestra valores RGB intentando centrar en el puntero
        if px > -1:
            cv2.putText(frame, '{}'.format(R), (px-45,py-5), font, 0.5, (0, 0, 255), 1)
            cv2.putText(frame, '     {}'.format(G), (px-54,py-5), font, 0.5, (0, 255, 0), 1)
            cv2.putText(frame, '         {}'.format(B), (px-54,py-5), font, 0.5, (255, 0, 0), 1)
            
        cv2.imshow('Cam', frame)   
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

```python
# Función que encuentra el píxel más brillante en una imagen.
def encontrar_pixeles(imagen):

    # Convertimos la imagen en una escala de grises para simplificar la búsqueda del píxel más brillante
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    # Encontrar las coordenadas del valor máximo en la imagen en escala de grises
    (val_min, val_max, local_min, local_max) = cv2.minMaxLoc(gris)
    
    # local_max será la coordenada (x, y) del píxel más brillante y local_min será la coordenada (x, y) del pixel más oscuro
   
    return local_max, local_min
```

```python
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()
    
    if ret:
        # Encuentra el píxel más brillante y más oscuro en el fotograma
        pixel_briliante, pixel_oscuro = encontrar_pixeles(frame)
        
        # Dibuja un círculo en el píxel más brillante (verde)
        cv2.circle(frame, pixel_briliante, 10, (0, 255, 0), 2)  # Círculo verde
        
        # Dibuja un círculo en el píxel más oscuro (rojo)
        cv2.circle(frame, pixel_oscuro, 10, (0, 0, 255), 2)  # Círculo rojo
        
        # Muestra la imagen con los píxeles resaltados
        cv2.imshow('Video con el píxel más brillante y más oscuro', frame)
    
    # Detener si se presiona la tecla ESC (código ASCII 27)
    if cv2.waitKey(20) & 0xFF == 27:
        break

# Libera el objeto de captura y destruye las ventanas
vid.release()
cv2.destroyAllWindows()
```

```python
def encontrar_brillante_8x8(imagen, tam_bloque=8):

    # Pintamos de gris la imagen para encontrar facilmente los pixeles en una escala de grises
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    h, w = gris.shape
    
    # Inicializamos el valor máximo y la posición
    max_brilho = -1
    max_pos = (0, 0)
    
    # Recorremos la imagen en bloques de 8x8 píxeles
    for y in range(0, h, tam_bloque):
        for x in range(0, w, tam_bloque):
            # Definir la región 8x8
            bloque = gris[y:y+tam_bloque, x:x+tam_bloque]
            brillo_bloque = np.mean(bloque)
            
            # Comparamos el brillo promedio del bloque con el valor máximo encontrado
            if brillo_bloque > max_brilho:
                max_brilho = brillo_bloque
                max_pos = (x, y)
    
    return max_pos

def encontrar_oscuro_8x8(imagen, tam_bloque=8):
   
    # Pintamos de gris la imagen para encontrar facilmente los pixeles en una escala de grises
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    h, w = gris.shape
    
    # Inicializamos el valor mínimo y la posición
    min_brilho = 256  # Los valores en escala de grises van de 0 a 255
    min_pos = (0, 0)
    
    # Recorremos la imagen en bloques de 8x8 píxeles
    for y in range(0, h, tam_bloque):
        for x in range(0, w, tam_bloque):
            # Definir la región 8x8
            bloque = gris[y:y+tam_bloque, x:x+tam_bloque]
            brillo_bloque = np.mean(bloque)
            
            # Comparamos el brillo promedio del bloque con el valor mínimo encontrado
            if brillo_bloque < min_brilho:
                min_brilho = brillo_bloque
                min_pos = (x, y)
    
    return min_pos
```

```python
# Captura video desde la cámara
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()
    
    if ret:
        # Encuentra la zona más brillante y más oscura en bloques de 8x8
        zona_briliante = encontrar_brillante_8x8(frame, tam_bloque=8)
        zona_oscura = encontrar_oscuro_8x8(frame, tam_bloque=8)
        
        # Dibuja un rectángulo en la zona más brillante (verde)
        cv2.rectangle(frame, zona_briliante, 
                      (zona_briliante[0] + 20, zona_briliante[1] + 20), 
                      (0, 255, 0), 2)  # Rectángulo verde
        
        # Dibuja un rectángulo en la zona más oscura (rojo)
        cv2.rectangle(frame, zona_oscura, 
                      (zona_oscura[0] + 20, zona_oscura[1] + 20), 
                      (0, 0, 255), 2)  # Rectángulo rojo
        
        # Muestra la imagen con los rectángulos
        cv2.imshow('Video con la zona más brillante y más oscura', frame)
    
    # Detener si se presiona la tecla ESC (código ASCII 27)
    if cv2.waitKey(20) & 0xFF == 27:
        break

# Libera el objeto de captura y destruye las ventanas
vid.release()
cv2.destroyAllWindows()
```

Pop art
Salida de la webcam "inspirada" en Andy Warhol (Marilyn, https://temasycomentariosartepaeg.blogspot.com/p/autor-andy-warhol-1928-1987-titulo.html)

```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cámara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resolución para ocupar menos pantalla
w=int(w/2)
h=int(h/2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w) #En Mac no reacciona a estos comandos
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*2,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
bl = collage[h:h+h,0:w]
br = collage[h:h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frameIN = vid.read()

    #Menor tamaño
    frame = cv2.resize(frameIN, (int(w),int(h)),cv2.INTER_NEAREST)

    if ret:
        #Separamos canales
        r = frame[:,:,2]
        g = frame[:,:,1]
        b = frame[:,:,0]

        #Jugamos con los valores de los planos
        tl[:,:,0] = b
        tl[:,:,1] = g
        tl[:,:,2] = r

        tr[:,:,0] = 255 - r
        tr[:,:,1] = g
        tr[:,:,2] = b
        
        bl[:,:,0] = r
        bl[:,:,1] = 255 - b
        bl[:,:,2] = g

        br[:,:,0] = b
        br[:,:,1] = g
        br[:,:,2] = 255 - r
    
        # Muestra composición
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Transformando la imagen capturada en una colección de círculos, cuyo radio depende de la intensidad promedio de la zona

```python
vid = cv2.VideoCapture(0)

# Se define el número de celdas
ncells = 10
off = int(ncells/2)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Dimensiones originales
        h, w, c = frame.shape
        #Redimensiono teniendo en cuenta ncells
        down_frame = cv2.resize(frame, (int(w/ncells),int(h/ncells)),cv2.INTER_NEAREST)
        #Dimensiones reducidas
        h2, w2, c2 = down_frame.shape

        #Separamos canales
        r = down_frame[:,:,0]
        g = down_frame[:,:,1]
        b = down_frame[:,:,2]

        # Creamos imagen negra
        gris_up_frame = np.zeros((h2*ncells,w2*ncells,1), dtype = np.uint8)

        # recorre valores de la imagen reducida
        for y in range(0,h2):
            for x in range(0,w2):
                # La suma de los valores RGB define el radio del círculo
                rad = int ( (r[y,x] + g[y,x] + b[y,x])/(ncells*3*2))
                cv2.circle(gris_up_frame,(x*ncells+off, y*ncells + off), rad, (255,255,255), -1)

    
        # Muestra fotograma resultante
        cv2.imshow('Cam', gris_up_frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture(0)

# Dimensiones de la cámara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Fuerzo a mitad de resolución para ocupar menos pantalla
w = int(w / 2)
h = int(h / 2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w)  # En Mac no reacciona a estos comandos
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

# Imagen conjunta 1x2
collage = np.zeros((h, w * 2, 3), dtype=np.uint8)
left_img = collage[:, 0:w]
right_img = collage[:, w:w*2]

while True:
    # fotograma a fotograma
    ret, frameIN = vid.read()

    # Menor tamaño
    frame = cv2.resize(frameIN, (int(w), int(h)), cv2.INTER_NEAREST)

    if ret:
        # Izquierda: fotograma normal
        left_img[:, :] = frame

        #Separamos canales
        r = frame[:,:,2]
        g = frame[:,:,1]
        b = frame[:,:,0]


        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Asignamos al canal verde un valor basado en la escala de grises
        right_img[:,:,0] = np.zeros_like(gray_frame)  # Canal azul lo dejamos en 0
        right_img[:,:,1] = gray_frame  # Canal verde: usamos la escala de grises
        right_img[:,:,2] = np.zeros_like(gray_frame)  # Canal rojo lo dejamos en 0

        # Derecha: simulación de cámara térmica

        # Muestra composición
        cv2.imshow('Cam', collage)

    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break

# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()

```
